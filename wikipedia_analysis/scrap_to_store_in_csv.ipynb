```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_manufacturing_companies_by_revenue'
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
table = soup.find('table',class_='wikitable sortable plainrowheads')
```

```python
table = table.tbody
```

```python
rows = table.find_all('tr')

rows[0]
```

```python
cols = [i.text for i in rows[0].find_all('th')]
cols = [i.strip() for i in cols]
cols
```

```python
df = pd.DataFrame(columns=cols)
```

```python
rows[1].find_all('td')
```

```python
master_list = []
for i in range(1,len(rows)):
    tds = rows[i].find_all('td')
    
    vals = [td.text.strip() for td in tds]
    vals[4] = vals[4].replace('\xa0','')
    
    d = {}
    d['No'] = vals[0]
    d['Company'] = vals[1]
    d['Industry'] = vals[2]
    d['Revenue (by US$ million)'] = vals[3]
    d['Headquarters'] = vals[4]
    
    master_list.append(d)

master_list
    
    
```

```python
df = pd.DataFrame(master_list)
df
```

```python
df.to_csv('list_of_companies.csv',index=0)
```
