```python
survey_raw_df = pd.read_csv('/Users/alenjose/Desktop/data/stack_overflow/survey_results_public.csv')
```

```python
survey_raw_df['LanguageHaveWorkedWith']
```

```python
languages = survey_raw_df['LanguageHaveWorkedWith'].to_frame()
languages
```

```python
lst = []
for idx,row in languages['LanguageHaveWorkedWith'][languages['LanguageHaveWorkedWith'].notna()].iteritems():
    langs = row.split(';')

    for lang in langs:
        if lang not in lst:
            lst.append(lang)
            languages[lang] = False
        languages.loc[idx,lang] = True 

        
languages.head()
```
