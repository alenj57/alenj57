```python
lst = []
for idx,row in languages['LanguageWorkedWith'][languages['LanguageWorkedWith'].notna()].iteritems():
    langs = row.split(';')
#     print(idx,langs)
    for lang in langs:
        if lang not in lst:
            lst.append(lang)
            languages[lang] = False
        languages.loc[idx,lang] = True 

        
languages.head()
```

```python
c = set()
for i in list(languages['LanguageWorkedWith'].str.split(';')):
    try:
        for j in i:
            c.add(j)
    except:
        pass
        
c
```

```python
for lang in c:
    try:
            print(lang)
            languages[lang] = languages['LanguageWorkedWith'].str.contains(lang, na=False)
    except:
        continue
```
