```python
# dataset : https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
confirmed = pd.read_csv('covid19_confirmed.csv')
deaths = pd.read_csv('covid19_deaths.csv')
recovered = pd.read_csv('covid19_recovered.csv')
```

```python
confirmed.head()
```

```python
confirmed = confirmed.drop(['Province/State','Lat','Long'],axis=1)
deaths = deaths.drop(['Province/State','Lat','Long'],axis=1)
recovered = recovered.drop(['Province/State','Lat','Long'],axis=1)
```

```python
# it is clear that country is spread over, need to remove the duplicates and 
# combine the values
confirmed['Country/Region'].value_counts().head()
```

```python
confirmed = confirmed.groupby('Country/Region').aggregate('sum')
deaths = deaths.groupby('Country/Region').aggregate('sum')
recovered = recovered.groupby('Country/Region').aggregate('sum')
```

```python
confirmed.tail()
```

```python
confirmed = confirmed.T
deaths = deaths.T
recovered = recovered.T
```

```python
confirmed.tail()
```

```python
new_cases = confirmed.copy()
for day in range(1,len(confirmed)):
    new_cases.iloc[day] = confirmed.iloc[day] - confirmed.iloc[day-1]
```

```python
new_cases.tail()
```

```python
growth_rate = confirmed.copy()

for day in range(1,len(confirmed)):
    growth_rate.iloc[day] = (new_cases.iloc[day] / confirmed.iloc[day-1])*100
```

```python
growth_rate.tail()
```

```python
active_cases = confirmed.copy()

for day in range(0,len(confirmed)):
    active_cases.iloc[day] = confirmed.iloc[day] - deaths.iloc[day] - recovered.iloc[day]
    
active_cases.tail()
```

```python
overall_growth_rate = confirmed.copy()

for day in range(0,len(confirmed)):
    overall_growth_rate.iloc[day] = ((active_cases.iloc[day]-active_cases.iloc[day-1])/active_cases.iloc[day-1])*100
```

```python
death_rate = confirmed.copy()

for day in range(0,len(confirmed)):
    death_rate.iloc[day] = (deaths.iloc[day]/confirmed.iloc[day]) * 100

death_rate.tail()
```

```python
hospitilization_rate_estimate = 0.05

hospitilization_needed = confirmed.copy()

for day in range(0,len(confirmed)):
    hospitilization_needed.iloc[day] = active_cases.iloc[day] * hospitilization_rate_estimate

hospitilization_needed.tail()
```

# VISUALIZATION :

##  Total Confirmed Cases :

```python
countries = ['Italy','Australia','US','China','India','France']

ax = plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title('Covid-19 - Total Confirmed Cases By Country',color='yellow')

for country in countries:
    confirmed[country].plot(label=country)
    
plt.legend(loc='upper left')    
plt.show()
```


## Growth Rate, Country Wise :

```python
countries = ['Italy','Germany','US','China']

for c in countries:
    growth_rate[c][30:50].plot.bar()
    ax = plt.subplot()
    ax.set_facecolor('black')
    ax.figure.set_facecolor('#121212')
    ax.tick_params(axis='x',colors='white')
    ax.tick_params(axis='y',colors='white')
    ax.set_title(f'Covid-19 - Growth Rate of {c}',color='yellow')
    plt.show()
```

```python
simulated_growth_rate = 0.30
```

```python
dates = pd.date_range(start='10/08/20',periods=40,freq='D')
dates = pd.Series(dates)
dates = dates.dt.strftime('%m/%d/%Y')
```

```python
simulated = confirmed.copy()
simulated = simulated.append(pd.DataFrame(index=dates))
```

```python
simulated
```

```python
for day in range(len(confirmed),len(confirmed)+40):
    simulated.iloc[day] = simulated.iloc[day - 1] * (1+ simulated_growth_rate)
    
simulated.tail()
```

```python
ax = plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title(f'Covid-19 - Future Simulation of Italy',color='yellow')
simulated['Italy'].plot()
plt.show
```
