# defining the dataset

```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast.keys()
```

```python
target_names = breast.target_names
target_names
```

```python
data = breast.data
data.shape
```

```python
target = breast.target

target.shape
```

```python
target = target.reshape(569,1)
```

```python
# combine the data and target values
df = np.concatenate([data,target],axis=1)

df = pd.DataFrame(df)
df.head()
```

```python
# find the column names:
features = breast.feature_names
features
```

```python
# add the target column name
features = np.append(features,'target')
```

```python
df.columns = features
df.head()
```

```python
df['target'].replace({0.0:'Benign',1:'Malignant'},inplace=True)
df.head(3)
```

# normalization :

```python
from sklearn.preprocessing import StandardScaler
```

```python
# normalized data with mean=0 and std=1
data_norm = StandardScaler().fit_transform(data)
data_norm.shape
```

```python
np.mean(data_norm)
```

```python
np.std(data_norm)
```

# PCA :

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(data_norm)
data = pca.transform(data_norm)
```

```python
pca_df = pd.DataFrame(data = data,columns=['PC1','PC2'])
pca_df
```

```python
pca = PCA(n_components=4)
pca.fit(data_norm)
data = pca.transform(data_norm)

pca_df = pd.DataFrame(data = data,columns=['PC1','PC2','PC3','PC4'])

pca.explained_variance_ratio_
```

```python
comps = pca.components_
```

```python
ax = plt.gca()
img = ax.imshow(comps, cmap='seismic', vmin=-1, vmax=1)
ax.set_title('Component Matrix')
plt.colorbar(img, label="weight")
ax.set_yticks([0,1,2,3])
t = ax.set_yticklabels(["PC{}".format(i+1) for i in range(4)])
ax.set_xticks(list(range(0, len(features)-1)))
t = ax.set_xticklabels(features, rotation=90)

plt.show()
```
