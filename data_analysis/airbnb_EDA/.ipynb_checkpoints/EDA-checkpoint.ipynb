# How does the price of Airbnb listings vary by property type and location?

# What is the average availability of Airbnb listings by neighborhood and property type?


```python
df['has_availability'].value_counts()
```

```python
df['has_availability'] = df['has_availability'].apply(lambda x: 1 if x == 't' else 0)
```

```python
merged = pd.merge(df.groupby('neighbourhood')['has_availability'].sum().reset_index(),
          df.groupby('neighbourhood')['has_availability'].count().reset_index(),
            on='neighbourhood', how='inner')
merged['percent'] = round((merged['has_availability_x'] / merged['has_availability_y']) * 100,1)
merged.sort_values('has_availability_x',ascending=False)[['neighbourhood','percent']].head(5)
```

```python
merged = pd.merge(df.groupby('room_type')['has_availability'].sum().reset_index(),
          df.groupby('room_type')['has_availability'].count().reset_index(),
            on='room_type', how='inner')
merged = merged.rename(columns={'has_availability_x': 'available', 'has_availability_y': 'total_count'})
merged['percent_available'] = round((merged['available'] / merged['total_count']) * 100, 1)
merged.sort_values('available',ascending=False)
```

# Which amenities are most commonly offered in Airbnb listings?

```python
import ast
df['amenities'] = df['amenities'].apply(lambda x:ast.literal_eval(x)) 
```

```python
amenities_lst = []
_ = df['amenities'].apply(lambda x: [amenities_lst.append(i) for i in x])
```

```python
amenities_set = set()
for i in amenities_lst:
    amenities_set.add(i)
amenities_dict = {key: 0 for key in amenities_set}

for key in amenities_lst:
    amenities_dict[key] = amenities_dict[key] + 1
```

```python
pd.DataFrame(list(amenities_dict.items()), columns=['amenities', 'total_count'])\
    .sort_values(by='total_count',ascending=False).head(15)\
        .set_index('amenities')
```

```python
amenities_df = pd.DataFrame(list(amenities_dict.items()), columns=['amenities', 'count'])
top15 = amenities_df.sort_values(by='count', ascending=False).head(15)

# create figure and axis
fig, ax = plt.subplots()
sns.barplot(data=top15, x='amenities', y='count', ax=ax, linewidth=3)

for p in ax.patches:
    value = p.get_height() / df.shape[0]
    text = f'{round(value * 100,1)}'
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(text, (x, y), ha='center', va='bottom')

ax.set_ylim(top=top15['count'].max() * 1.4)


# set the x-tick labels to vertical
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

# show the plot
plt.tight_layout()
plt.show()
```

# How do the review scores for cleanliness, check-in, communication, and value vary by property type and host characteristics?

# How does the Airbnb listings change over time?

```python
df['host_since'] = pd.to_datetime(df['host_since'])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

df_count = df.groupby(pd.Grouper(key='host_since', freq='M'))['id'].count()

# Reset the index and rename the columns
df_count = pd.DataFrame(df_count.reset_index(name='count'))

sns.lineplot(data=df_count, x='host_since', y='count', ax=ax)

plt.show()
```

```python
df_year_count = df.groupby(pd.Grouper(key='host_since', freq='Y'))['id'].count().reset_index()
df_year_count['host_since'] = df_year_count['host_since'].dt.strftime('%Y')

fig, ax = plt.subplots(figsize=(8, 5))

sns.barplot(data=df_year_count, x='host_since', y='id', ax=ax)

plt.xlabel('Year')
plt.ylabel('New Hosts')
plt.title('New Hosts Over the years')

plt.tight_layout()
plt.show()
```

**The data shows the number of new hosts who joined Airbnb each year from 2008 to 2022. 
The highest number of new hosts was recorded in 2013, with 1052 new hosts joining the platform. 
The number of new hosts has since decreased, with 220 new hosts joining in 2021 and 242 in 2022. 
The period between 2014 and 2018 saw relatively stable numbers of new hosts joining each year, 
while there was a sharp decline in 2019 before a slight increase in 2020. Overall, the data suggests 
that Airbnb's growth in terms of new hosts joining the platform has slowed in recent years.**

# What are the most common reasons cited by guests for giving low review scores, and are these reasons more prevalent in certain neighborhoods or property types?

```python
df1 = pd.read_csv('/Users/alenjose/Desktop/data/airbnb_EDA/reviews.csv')
```

```python
import nltk
nltk.download('stopwords')
import string

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
punctuation = string.punctuation
```

```python
def filter_func(x):
    try:
        tokens = nltk.word_tokenize(x)
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]    
    except:
        filtered_tokens = 'none'
    return filtered_tokens

df1['filtered'] = df1['comments'].swifter.apply(lambda x:filter_func(x))
```

```python
def remove_punc(x):
    try:
        lst = [word for word in x if word not in punctuation]
    except:
        lst = 'none'
    return lst
```

```python
from textblob import TextBlob

def get_sentiment(x):
    blob = TextBlob(" ".join(remove_punc(x)))
    sentiment = blob.sentiment
    return sentiment.polarity,sentiment.subjectivity

df1['filtered'] = df1['filtered'].swifter\
     .apply(lambda x:get_sentiment(x))
```

```python
df1['polarity'] = df1['filtered'].apply(lambda x:x[0])
df1['subjectivity'] = df1['filtered'].apply(lambda x:x[1])
df1.drop('filtered',axis=1,inplace=True)
df1 = df1.groupby('listing_id')[['polarity','subjectivity']].mean()
```

```python
merged = pd.merge(df[['id','review_scores_rating']],
         df1,
         left_on='id',
         right_on='listing_id',
         how='inner')
merged['polarity'] = merged['polarity'].apply(lambda x:x*100)
merged
```

```python
sns.scatterplot(x='polarity',y='review_scores_rating',data=merged)
plt.show()
```
