```python
import pandas as pd
%pylab inline
import seaborn 
```

```python
data = pd.read_csv('uber-raw-data-apr14.csv')
```

```python
data.head()
```

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
```

```python
data['Date/Time'][0]
```

```python
# here apply and map both works. 

def get_day(dt):
    return dt.day

def get_weekday(dt):
    return dt.weekday()

def get_hour(dt):
    return dt.hour

data['day'] = data['Date/Time'].map(get_day)
data['weekday'] = data['Date/Time'].map(get_weekday)
data['hour'] = data['Date/Time'].map(get_hour)
```

```python
# so there are no calls at the hour 24
data[data['hour']==24]
```

```python
data.tail()
```

# Analysis :

```python
days = list(range(1,31))
data['day'].hist(bins=30,rwidth=.8, range=(0.5,30.5),figsize=(15,10))
xlabel('date of the month')
ylabel('Frequency')
title('Frequency By Day - Uber - Apr 2014')
xticks(days,rotation='vertical',size=9)
plt.show()
```

```python
# for k,rows in data.groupby('day'):
#     print(k, len(rows))
def get_count(df):
    return len(df)

by_date = data.groupby('day').apply(get_count)
by_date
```

```python
by_date.plot(kind='bar')
```

```python
by_freq_sorted = by_date.sort_values()
by_freq_sorted.plot(kind='bar')
```

```python
hours = list(range(1,25))
data['hour'].plot(kind='hist',bins=24,rwidth=.8,range=(0.5,24.5))
xticks(hours)
;
```

```python
days = ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
data['weekday'].plot(kind='hist',bins=6,rwidth=.8,range=(-.5,5.5),color='green',alpha=.6,figsize=(8,5))
xlabel('weekdays')
title('Frequency by weekday')
xticks(range(0,6),days)
plt.show()
```

## Cross Analysis :

```python
cross = data.groupby('weekday hour'.split()).apply(get_count).unstack()
cross
```

```python
seaborn.heatmap(cross)
```
