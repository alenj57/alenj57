```python
#plotting stock price trend
ax1 = stock['Prices'].plot(style=['-'])
plt.title("Stock Prices")

ax1.legend()
plt.plot()
```

```python
#plotting return
ax2 = stock['Returns'].plot(style=['-'])
plt.title("Returns")

ax2.legend()
plt.plot()
```

```python
# Plot both
plt.plot(stock['Prices'])
plt.title('Stock Price')
plt.ylabel('Price');
plt.show()

plt.plot(stock['Returns'],'r')
plt.title('Stock Return')
plt.ylabel('Return');
plt.show();
```

```python
#dotted plot

stock['Prices'].plot(style='k.')
plt.show()
```

```python
#dotted plot
stock['Returns'].plot(style='k.')
plt.show()
```

```python
#Histogram of prices
stock['Prices'].hist()
plt.show()
```

```python
#Histogram of Returns
#Histogram
stock['Returns'].hist()
plt.show()
```

```python
#Kernel Density plot

stock['Prices'].plot(kind='kde')
plt.show()

stock['Returns'].plot(kind='kde')
plt.show()
```

```python
#plotting moving average (rolling mean of Prices)
price_mean = stock['Prices'].rolling(20,win_type=None).mean()
price_mean.plot(style=['-', '--', ':'])
```

```python
#plotting moving average (rolling mean of Returns)
price_mean = stock['Returns'].rolling(30,win_type=None).mean()
price_mean.plot(style=['-', '--', ':'])
```

```python
#Exponetial smoothing

from statsmodels.tsa.holtwinters import SimpleExpSmoothing

model = SimpleExpSmoothing(stock['Prices']).fit(smoothing_level=.1,optimized=True)

stock['Prices'].plot(marker='o', color='red')
model.fittedvalues.plot(marker='o', color='blue')
```

```python
#Lag plot - it is used to check if the series is random or not
from pandas.plotting import lag_plot
lag_plot(stock['Prices'])
```

```python
#exhibits a linear pattern.
#This shows that the data are strongly non-random and further suggests that an autoregressive model might be appropriate.
```

```python
#Autocorrelation plot

from pandas.plotting import autocorrelation_plot
autocorrelation_plot(stock['Prices'])
```

```python
#Partial Autocorrelation

from pandas import Series
from statsmodels.graphics.tsaplots import plot_pacf

#pacf
plot_pacf(stock['Prices'], lags=5)
```
