```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import EarlyStopping

import nltk
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
stop_words = list(stopwords.words('english'))

from tensorflow.keras.models import load_model
import tensorflow as tf
```

```python
df = pd.read_csv('/Users/alenjose/Desktop/data/project_data/training_data/cleaned_model_data.csv',index_col=0)
```

```python
df['tweet_text_sw'] = df['tweet_text']
```

```python
white_tok = WhitespaceTokenizer()
lemmat = WordNetLemmatizer()
stemmer = PorterStemmer()

def lemmatize_and_stem_text(text):
    string = ""
    for word in white_tok.tokenize(text):
        lemmatized_word = lemmat.lemmatize(word)
        stemmed_word = stemmer.stem(lemmatized_word)
        string = string + stemmed_word + " "
    return string

df['tweet_text_sw'] = df.tweet_text_sw.apply(lemmatize_and_stem_text)

# 0- negative 1-positive
df['target'] = df['target'].map({0:0,4:1})
```

```python
# 80 - 20 split (training and testing)
x_train, x_test_temp, y_train, y_test_temp = train_test_split(df['tweet_text_sw'], df['target'], test_size=0.2, random_state=42)

# 50 - 50 split on temp set
x_val, x_test, y_val, y_test = train_test_split(x_test_temp, y_test_temp, test_size=0.5, random_state=42)
```

### Testing out with different parameters 

```python
train_sentences = x_train.astype(str).tolist()
train_labels = y_train.values

val_sentences = x_val.astype(str).tolist()
val_labels = y_val.values

# hyperparameters
vocab_size = 10000
oov_tok = '<OOV>'
embedding_dim = 128
max_length = 250
padding_type='post'
trunc_type='post'

tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index
train_sequences = tokenizer.texts_to_sequences(train_sentences)
train_padded = pad_sequences(train_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
val_sequences = tokenizer.texts_to_sequences(val_sentences)
val_padded = pad_sequences(val_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
```

```python
# model initialization
model = keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    keras.layers.Bidirectional(keras.layers.LSTM(128)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
# compile model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()

print('\n')

early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

num_epochs = 8
history = model.fit(train_padded, train_labels, 
                    epochs=num_epochs, verbose=1, 
                    validation_split=0.1,
                   callbacks=[early_stop])
```

```python
# accuracy and loss:
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

best_epoch = 3

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# training vs validation acc
ax1.plot(train_acc, label='Training Accuracy')
ax1.plot(val_acc, label='Validation Accuracy')
ax1.set_title('Training vs Validation Accuracy')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)
ax1.axvline(x=best_epoch-1, color='red', linestyle='--')  
ax1.annotate('Best Epoch', xy=(best_epoch-1, 0.8), xytext=(best_epoch-1, 0.8), color='red', arrowprops=dict(facecolor='red', shrink=0.05))

# training vs validation loss
ax2.plot(train_loss, label='Training Loss')
ax2.plot(val_loss, label='Validation Loss')
ax2.set_title('Training vs Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)
ax2.axvline(x=best_epoch-1, color='red', linestyle='--') 
ax2.annotate('Best Epoch', xy=(best_epoch-1, 0.39), xytext=(best_epoch-1, 0.4), color='red')

plt.tight_layout()
plt.show()
```

**Implications:**

    Embedding Dimension: Only made a slight boost but at the cost of increased training time.

    LSTM Units: Marginal improvement in performance, but higher training time. 

    Dense Units: No significant difference.

    Sequence Length: Didnt significantly affect the model performance.

    Vocabulary Size: No change in results.

### Training with a reduced learning rate and keeping all the original params: 

```python
# resetting all other parameters to default for faster execution
vocab_size = 7000 
embedding_dim = 100
max_length = 200 

# tokenize sentences
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index

# convert train dataset to sequence and pad sequences
train_sequences = tokenizer.texts_to_sequences(train_sentences)
train_padded = pad_sequences(train_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)

# convert test dataset to sequence and pad sequences
val_sequences = tokenizer.texts_to_sequences(val_sentences)
val_padded = pad_sequences(val_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
```

```python
from tensorflow.keras.optimizers import Adam

new_model = keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    keras.layers.Dense(24, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# new model compiled
new_model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0001), metrics=['accuracy'])
```

```python
# model summary
new_model.summary()

print('\n')

early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

num_epochs = 8
history_2 = new_model.fit(train_padded, train_labels, 
                    epochs=num_epochs, verbose=1, 
                    validation_split=0.1,
                   callbacks=[early_stop])
```

```python
# epochs
epochs = [1, 2, 3, 4, 5, 6, 7, 8]

train_acc = history_2.history['accuracy']
val_acc = history_2.history['val_accuracy']
train_loss = history_2.history['loss']
val_loss = history_2.history['val_loss']

# best epoch
best_epoch = 8

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))


# training vs validation acc
ax1.plot(train_acc, label='Training Accuracy')
ax1.plot(val_acc, label='Validation Accuracy')
ax1.set_title('Training vs Validation Accuracy')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)
ax1.axvline(x=best_epoch-1, color='red', linestyle='--')  
ax1.annotate('Best Epoch', xy=(best_epoch-1, 0.8), xytext=(best_epoch-1, 0.8), color='red')

# training vs validation loss
ax2.plot(train_loss, label='Training Loss')
ax2.plot(val_loss, label='Validation Loss')
ax2.set_title('Training vs Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)
ax2.axvline(x=best_epoch-1, color='red', linestyle='--') 
ax2.annotate('Best Epoch', xy=(best_epoch-1, 0.39), xytext=(best_epoch-1, 0.4), color='red')

plt.tight_layout()
plt.show()
```

## It is clear that no further training is needed, the validation loss has flattened out and the validation accuracy is only incrementing in smaller steps

## Add a dropout layer to the keras model to prevent overfitting : 

```python
from tensorflow.keras.layers import Dropout

drop_model = keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    Dropout(0.5),  # Dropout layer after Embedding
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    Dropout(0.5),  # Dropout layer after LSTM
    keras.layers.Dense(24, activation='relu'),
    Dropout(0.5),  # Dropout layer before final layer
    keras.layers.Dense(1, activation='sigmoid')
])

drop_model.compile(loss='binary_crossentropy', 
                   optimizer='adam', 
                   metrics=['accuracy'])

drop_model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

num_epochs = 8
history = drop_model.fit(train_padded, train_labels, 
                    epochs=num_epochs, verbose=1, 
                    validation_split=0.1,
                   callbacks=[early_stop])
```

```python
epochs3 = [1, 2, 3, 4, 5, 6, 7, 8]

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# best epoch 
best_epoch = 5

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# training vs validation acc
ax1.plot(train_acc, label='Training Accuracy')
ax1.plot(val_acc, label='Validation Accuracy')
ax1.set_title('Training vs Validation Accuracy')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)
ax1.axvline(x=best_epoch-1, color='red', linestyle='--')  
ax1.annotate('Best Epoch', xy=(best_epoch-1, 0.8), xytext=(best_epoch-1, 0.8), color='red')

# training vs validation loss
ax2.plot(train_loss, label='Training Loss')
ax2.plot(val_loss, label='Validation Loss')
ax2.set_title('Training vs Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)
ax2.axvline(x=best_epoch-1, color='red', linestyle='--') 
ax2.annotate('Best Epoch', xy=(best_epoch-1, 0.5), xytext=(best_epoch-1, 0.4), color='red')

plt.tight_layout()
plt.show()
```
