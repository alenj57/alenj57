```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import keras
from keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split

from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import EarlyStopping

import nltk
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
stop_words = list(stopwords.words('english'))

from tensorflow.keras.models import load_model
import tensorflow as tf
```

```python
df = pd.read_csv('/Users/alenjose/Desktop/data/project_data/training_data/cleaned_model_data.csv',index_col=0)
```

```python
df['tweet_text_sw'] = df['tweet_text']
```

```python
df['tweet_text'] = df['tweet_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
print(len(df))
print('there are now',(len(df) - len(df[df['tweet_text'].notna() & (df['tweet_text'] != '')])),'empty texts after stopwords removal')
```

## https://www.analyticsvidhya.com/blog/2022/01/sentiment-analysis-with-lstm/

## without stop words : 

```python
white_tok = WhitespaceTokenizer()
lemmat = WordNetLemmatizer()
stemmer = PorterStemmer()

def lemmatize_and_stem_text(text):
    string = ""
    for word in white_tok.tokenize(text):
        lemmatized_word = lemmat.lemmatize(word)
        stemmed_word = stemmer.stem(lemmatized_word)
        string = string + stemmed_word + " "
    return string

df['tweet_text'] = df.tweet_text.apply(lemmatize_and_stem_text)

df.head()
```

### Visualize Word Frequency : 

```python
from collections import Counter

plt.rcParams.update({'font.size': 14})

# all tweets
all_sentences = df['tweet_text'].tolist()

# tokenizing the sentences
tokens = [word for sentence in all_sentences for word in sentence.split()]

# frequency of each token
frequency = Counter(tokens)

# sorting tokens based on freq
sorted_tokens = sorted(frequency.items(), key=lambda x: x[1], reverse=True)

# subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# word freq
ax1.plot([x[1] for x in sorted_tokens])
ax1.axvline(x=7000, color='red', linestyle='dashed')
ax1.set_title("Word Frequencies", fontsize=16)
ax1.set_xlabel("Word Rank", fontsize=15)
ax1.set_ylabel("Frequency", fontsize=15)

# cumulative sum (word freq)
cumulative_freq = np.cumsum([x[1] for x in sorted_tokens])

# cumulative coverage
cumulative_coverage = cumulative_freq / len(tokens)

# cumulative coverage of words
ax2.plot(cumulative_coverage)
ax2.axhline(y=0.9, color='red', linestyle='dashed')
ax2.axvline(x=7000, color='green', linestyle='-')
ax2.set_title("Cumulative Coverage of Words", fontsize=16)
ax2.set_xlabel("Number of Words", fontsize=15)
ax2.set_ylabel("Coverage", fontsize=15)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

**From the above plot it is clear that 7000 which intersects the 90% coverage line covers most of the vocabulary**

```python
# 0- negative 1-positive
df['target'] = df['target'].map({0:0,4:1})

# 80 - 20 split (training and testing)
x_train, x_test_temp, y_train, y_test_temp = train_test_split(df['tweet_text'], df['target'], test_size=0.2, random_state=42)

# 50 - 50 split on the ttemp data
x_val, x_test, y_val, y_test = train_test_split(x_test_temp, y_test_temp, test_size=0.5, random_state=42)
```

```python
# extract tweets and labels for training and testing
train_sentences = x_train.astype(str).tolist()
train_labels = y_train.values

val_sentences = x_val.astype(str).tolist()
val_labels = y_val.values

# hyperparameters
vocab_size = 7000 
oov_tok = '<OOV>'
embedding_dim = 100
max_length = 200 
padding_type='post'
trunc_type='post'

# tokenize sentences
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index

# convert train,test datasets to sequence and pad sequences
train_sequences = tokenizer.texts_to_sequences(train_sentences)
train_padded = pad_sequences(train_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
val_sequences = tokenizer.texts_to_sequences(val_sentences)
val_padded = pad_sequences(val_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)


# model initialise
model = keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    keras.layers.Dense(24, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
# compile model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
# model summary
model.summary()
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

num_epochs = 5
history = model.fit(train_padded, train_labels, 
                    epochs=num_epochs, verbose=1, 
                    validation_split=0.1,
                   callbacks=[early_stop])
```

```python
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

best_epoch = 2  

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# training, validation loss
ax1.plot(epochs, train_loss, label='Training Loss', marker='o')
ax1.plot(epochs, val_loss, label='Validation Loss', marker='o')
ax1.axvline(x=best_epoch, color='r', linestyle='--')
ax1.set_title('Training and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

# training, validation accuracy 
ax2.plot(epochs, train_accuracy, label='Training Accuracy', marker='o')
ax2.plot(epochs, val_accuracy, label='Validation Accuracy', marker='o')
ax2.axvline(x=best_epoch, color='r', linestyle='--')
ax2.set_title('Training and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

## With stop words : 

```python
df['tweet_text_sw'] = df.tweet_text_sw.apply(lemmatize_and_stem_text)
```

```python
# 80 - 20 split (training and testing)
x_train, x_test_temp, y_train, y_test_temp = train_test_split(df['tweet_text_sw'], df['target'], test_size=0.2, random_state=42)

# 50 - 50 split on temp data
x_val, x_test, y_val, y_test = train_test_split(x_test_temp, y_test_temp, test_size=0.5, random_state=42)
```

```python
train_sentences = x_train.astype(str).tolist()
train_labels = y_train.values

val_sentences = x_val.astype(str).tolist()
val_labels = y_val.values

# hyperparameters
vocab_size = 7000 
oov_tok = '<OOV>'
embedding_dim = 100
max_length = 200 
padding_type='post'
trunc_type='post'

# tokenize sentences
tokenizer = Tokenizer(num_words = vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(train_sentences)
word_index = tokenizer.word_index

# convert train, test dataset to sequence and pad sequences
train_sequences = tokenizer.texts_to_sequences(train_sentences)
train_padded = pad_sequences(train_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
val_sequences = tokenizer.texts_to_sequences(val_sentences)
val_padded = pad_sequences(val_sequences, padding=padding_type, maxlen=max_length, truncating=trunc_type)
```

```python
# model initialization
model = keras.Sequential([
    keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    keras.layers.Dense(24, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
# compile model
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
# model summary
model.summary()

print('\n')

early_stop = EarlyStopping(monitor='val_loss', patience=3, verbose=1, restore_best_weights=True)

num_epochs = 8
history = model.fit(train_padded, train_labels, 
                    epochs=num_epochs, verbose=1, 
                    validation_split=0.1,
                   callbacks=[early_stop])
```

```python
# accuracy and loss values:
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

best_epoch = 3

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# plotting train vs val acc
ax1.plot(train_acc, label='Training Accuracy')
ax1.plot(val_acc, label='Validation Accuracy')
ax1.set_title('Training vs Validation Accuracy')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)
ax1.axvline(x=best_epoch-1, color='red', linestyle='--')  

# plotting train vs val loss
ax2.plot(train_loss, label='Training Loss')
ax2.plot(val_loss, label='Validation Loss')
ax2.set_title('Training vs Validation Loss')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)
ax2.axvline(x=best_epoch-1, color='red', linestyle='--') 

plt.tight_layout()
plt.show()
```
