## Although word cloud is controversial as it leaves out viewers with little to no context. But when it comes to tweets, textual analysis is very important and to know the words frequently used gives us a general idea.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Let's assume that df is your DataFrame and 'tweet_without_username' is the column with the tweets
tweets_text = df['tweet_text_with_sw'].str.cat(sep=' ')

# Generate a word cloud image
wordcloud = WordCloud(background_color="white", max_words=1000).generate(tweets_text)

# Display the generated image with matplotlib 
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show() 
```

```python
import nltk
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Download the NLTK stop words
nltk.download('stopwords')
from nltk.corpus import stopwords

# Combine NLTK stop words with WordCloud's STOPWORDS
stop_words = set(stopwords.words('english')).union(set(STOPWORDS))

# Join all the text in the 'tweet_without_username' column
tweets_text = df['tweet_text_with_sw'].str.cat(sep=' ')

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stop_words, background_color="white", max_words=1000).generate(tweets_text)

# Display the generated image with matplotlib 
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Zipf's law states that the frequency of the word is inversely proportional to its rank in the frequency table. That is, the more frequent word will occur approximately twice as often as the second most frequent word.

```python
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer

corpus = df['tweet_text_with_sw']

vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(corpus)

# Sum up the counts of each vocabulary word
sum_words = X.sum(axis=0)

# Get word frequencies
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)

# Create a DataFrame from the words_freq list of tuples
word_freq_df = pd.DataFrame(words_freq, columns=['word', 'freq'])

word_freq_df.head()
```

```python
import numpy as np

# create ranks
word_freq_df['rank'] = word_freq_df['freq'].rank(method='min', ascending=False)
plt.figure(figsize=(8,6))
plt.scatter(np.log(word_freq_df['rank']), np.log(word_freq_df['freq']))
plt.xlabel('Log Rank')
plt.ylabel('Log Frequency')
plt.title('Log-Log plot of Word Frequencies')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# dataframe with ranks
word_freq_df = word_freq_df.sort_values(by='freq', ascending=False).reset_index()
word_freq_df['rank'] = word_freq_df.index + 1

# ideal zipf distribution (formula: `frequency = constant/rank`)
word_freq_df['ideal_freq'] = word_freq_df['freq'][0] / word_freq_df['rank']

# Generate log-log plot
plt.figure(figsize=(10,6))
plt.loglog(word_freq_df['rank'], word_freq_df['freq'], label='Actual')
plt.loglog(word_freq_df['rank'], word_freq_df['ideal_freq'], label='Ideal Zipf', linestyle='--')
plt.title('Word Frequencies')
plt.xlabel('Rank')
plt.ylabel('Frequency')
plt.legend()

# show the top 3 words (frequency)
for i in range(3):
    plt.annotate(word_freq_df.iloc[i]['word'], 
                 (word_freq_df.iloc[i]['rank'], word_freq_df.iloc[i]['freq']))


plt.show()
```

### The higher ranked words at the left on the x-axis have a higher frequency on the y-axis and towards the left of the x-axis as the rank decreases so does the frequency of that term. Although the law dictates that the actual observation will follow a near-Zipfian line, it is interesting to note that there is a chance that the tweets use frequent words more heavily than the other text corpora.
